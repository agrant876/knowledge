```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data
exports = pd.read_csv('data/country/jamaica/exports.csv')
exports
```

```python
imports = pd.read_csv('data/country/jamaica/imports.csv')
imports
```

```python
# make a square figure and axes
plt.figure(1, figsize=(8,8))
ax = plt.axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = exports['Unnamed: 0'][1:11]

patches, texts, numbers = plt.pie(exports['2013'][1:11],
                autopct='%1.1f%%', labeldistance=1.5, startangle=90)
                
plt.legend(patches,labels,bbox_to_anchor=(1.5, 1))
                
plt.show()

```

```python
# make a square figure and axes
plt.figure(1, figsize=(8,8))
ax = plt.axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = imports['Imports'][1:11]

patches, texts, numbers = plt.pie(imports['2014'][1:11],
                autopct='%1.1f%%', labeldistance=1.5, startangle=90)
                
plt.legend(patches,labels,bbox_to_anchor=(1.5, 1))
                
plt.show()
```

```python
fig, ax = plt.subplots()
ind = np.arange(10)
width = 0.35

ax.bar(ind,exports['2013'][1:11],width)
ax.bar(ind+width,imports['2013'][1:11],width,color='r')

ax.set_xticks(ind + width)
ax.set_xticklabels(ind)
          
plt.show()

imports['Imports'][1:11]
```
